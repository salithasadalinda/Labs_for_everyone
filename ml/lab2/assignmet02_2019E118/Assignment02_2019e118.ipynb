# Feature Selection

1. Use Boston housing data for this assignment. Load the data and extract
the feature matrix and target vector separately.






```python
import numpy as np
import pandas as pd

# import the data set

data = pd.read_csv('Boston.csv')

#extract the feature matrix and target vector
features = data.drop('medv', axis=1).values

target = data['medv'].values

```

```python
print("features: \n",features)
```

```python
print("target values: \n",target)
```

2. Use the Person correlation to find the correlation between features.
Draw the heatmap to show this correlation.

```python
# libraries
import seaborn as sns
import matplotlib.pyplot as plt
# Person correlation to find the correlation between features

corr_matrix = np.corrcoef(features.T)
sns.heatmap(corr_matrix, annot=True, fmt=".2f", xticklabels=data.columns[:-1], yticklabels=data.columns[:-1], cmap='YlGnBu')
plt.title("Correlation Heatmap")
plt.show()
```

3. Extract the correlation of all variables with the target variable.

```python
corr_with_target = np.abs(corr_matrix[-1, :-1])
```

4. Display the first three features with high correlation and state the
threshold value for this selection.

```python
# first three features with high correlation
threshold = 0.5
high_corr_features = data.columns[:-1][np.argsort(corr_with_target)[::-1]][:3]
print("high correlation features :", high_corr_features)
print("Threshold :", threshold)
```

5. Now use Lasso regression and find the number of selected and eliminated variables.

```python
# lasso model
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)  # parameter alpha
lasso.fit(features, target)
num_selected = np.sum(lasso.coef_ != 0)
num_eliminated = np.sum(lasso.coef_ == 0)
print("#selected variables:", num_selected)
print("#eliminated variables:", num_eliminated)
```

6. Find the importance of each feature and plot the values using a bar
graph.

```python
feature_importance = np.abs(lasso.coef_)
plt.bar(range(len(feature_importance)), feature_importance, tick_label=data.columns[:-1], color='green')
plt.title("The importance of features")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.xticks(rotation=90)
plt.show()
```
